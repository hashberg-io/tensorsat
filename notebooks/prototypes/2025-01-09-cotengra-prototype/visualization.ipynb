```python
import matplotlib.pyplot as plt
import cotengra as ct
from _sat import CNFInstance
```

# Visualizations

```python
cnf_instance = CNFInstance.random(k=3, n=3, m=4, rng=0)
cnf_instance
```

```python
network = cnf_instance.network()
network.wiring.hypergraph.plot(
    figsize=(5, 5),
    draw_edge_labels=True,
    hyperedge_style="solid",
);
```

```python
network.wiring.hypergraph.plot(
    figsize=(5, 5),
    hyperedge_style="solid",
    node_color="centrality",
    edge_color=False,
    draw_edge_labels=False,
);
```

```python
t = network.wiring.contraction_tree
t.plot_flat(figsize=(8, 8));
```

```python
t.plot_rubberband(
    figsize=(8, 8),
    hyperedge_style="solid",
);
```

```python
t.plot_tent(
    figsize=(4, 8),
    colorbars=True,
    raw_edge_alpha=0.1,
);
```

```python
t.plot_ring(
    figsize=(6, 8),
    colorbars=True,
    raw_edge_alpha=0.1,
);
```

```python
print(f"{t.total_flops() = }")  # total number of multiplications (total cost)
print(f"{t.total_write() = }B") # total memory written (total write)
print(f"{t.max_size() = }B")    # max size of intermediate tensor (write)
print(f"{t.peak_size() = }B")   # max concurrent size of intermediate tensors (peak)
t.plot_contractions(figsize=(10, 5));
```
